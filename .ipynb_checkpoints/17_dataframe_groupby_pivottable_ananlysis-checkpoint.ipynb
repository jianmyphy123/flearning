```python
candidate_to_votes_s = primary.groupby('candidate')['votes'].sum().sort_values()
```

```python
state_party_to_votes_s = primary.groupby(['state', 'party'])['votes'].sum()
```

```python
state_party_to_votes_s.head()
```

```python
state_to_votes_s = primary.groupby('state')['votes'].sum()
```

```python
state_to_votes_s.head()
```

```python
state_party_to_votes_pcts_s = state_party_to_votes_s / state_to_votes_s
```

```python
state_party_to_votes_pcts_s.head()
```

```python
state_party_to_votes_pcts_s.unstack().plot(kind='barh', stacked=True, fontsize=8)
```

```python
func = lambda agg_df: agg_df.sort_values('votes', ascending=False).iloc[0]
```

```python
primary.head()
```

```python
winners = primary.groupby('fips').agg(func)
```

```python
winners.head()
```

```python
countries.head()
```

```python
countries.columns
```

```python
countries['RHI825214'].head()
```

```python
countries[['fips', 'RHI825214']].head()
```

```python
winners.head()
```

```python
winnsers_county_races = pd.merge(winners, countries[['fips', 'RHI825214']],
                                 left_index=True, right_on='fips', how='left')
```

```python
winnsers_county_races.head()
```

```python
winnsers_county_races = winnsers_county_races.rename(columns={'RHI825214': 'white_pcts'})
```

```python
winnsers_county_races.head()
```

```python
winners_county_white_pcts = winnsers_county_races.groupby(['party', 'candidate'])['white_pcts'].mean()
```

```python
winners_county_white_pcts.head()
```

```python
ax = winners_county_white_pcts.plot(kind='barh', fontsize=8)
```

```python
ax.set_xlim([50, 100])
plt.tight_layout()
```
