```python
battles = battles.set_index(['name'])
```

```python
battles
```

```python
battles.head()
```

```python
large_battle_mask = battles['attacker_size'] + battles['defender_size'] > 10000
```

```python
large_battle = battles.loc[large_battle_mask, ['attacker_size', 'defender_size']]
```

```python
large_battle.head()
```

```python
ax2 = large_battle.plot(kind='barh', stacked=True, fontsize=8)
```

```python
large_battle['attacker_pcts'] = large_battle['attacker_size'] / (large_battle['attacker_size'] + large_battle['defender_size'])
```

```python
large_battle['defender_pcts'] = \
    large_battle['defender_size'] / (large_battle['attacker_size'] + large_battle['defender_size'])
```

```python
ax3 = large_battle[['attacker_pcts', 'defender_pcts']].plot(kind='barh', stacked=True, fontsize=8)
```

```python
col_names = battles.columns[4:12]
```

```python
house_names = battles[col_names].fillna('None')
```

```python
house_names[:5]
```

```python
house_names = np.unique(house_names)
```

```python
house_names
```

```python
house_names = house_names[house_names != 'None']
```

```python
house_names
```

```python
houses_to_battle_counts = pd.Series(0, index=house_names)
```

```python
houses_to_battle_counts
```

```python
battles.head()
```

```python
for col in col_names:
    houses_to_battle_counts = houses_to_battle_counts.add(battles[col].value_counts(), fill_value=0)
```
